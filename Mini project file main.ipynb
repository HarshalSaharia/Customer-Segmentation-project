```python
#creating clusters based on age and spending score
X1 = customer_dataset.loc[:,['Age','Spending Score (1-100)']].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
        kmeans = KMeans(n_clusters = k, init = 'k-means++')
        kmeans.fit(X1)
        wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth = 2, color = 'red', marker = '8')
plt.plot('Elbow Method')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
label = kmeans.fit_predict(X1)

print(label)

print(kmeans.cluster_centers_)

# Visualize our clusters(basically different groups):

plt.scatter(X1[:,0],X1[:,1], c=kmeans.labels_,cmap = 'rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'black')
plt.title('KMeans for Age and spending ')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
X3 = customer_dataset.loc[:,['Age','Annual Income (k$)','Spending Score (1-100)']].values
kmeans = KMeans(n_clusters = 5)
label = kmeans.fit_predict(X3)

print(label)

print(kmeans.cluster_centers_)

# visualize clusters:

plt.scatter(X3[:,0],X3[:,1], c=kmeans.labels_,cmap = 'rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'black')
plt.title('KMeans clustering for Age and Annual Income')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.show()
```
